### O que é melt?

O pandas.melt() é usado para transformar um DataFrame de formato “largo” (wide) para um formato “longo” (long).

_Em termos simples_:
Ele desfaz as colunas e as transforma em linhas, mantendo apenas as colunas que você quiser fixas.

**Parâmetros principais do melt:**

| Parâmetro    | O que faz                                                      |
| ------------ | -------------------------------------------------------------- |
| `id_vars`    | As colunas que você quer manter fixas (ex: país, tipo de café) |
| `value_vars` | As colunas que você quer derreter (geralmente anos)            |
| `var_name`   | Nome da nova coluna que guardará os nomes das antigas colunas  |
| `value_name` | Nome da nova coluna com os valores                             |


```python
# Vamos transformar o dataset de produção para formato longo
prod_long = df_production.melt(
    id_vars=['Country', 'Coffee type', 'Total_production'], # São as colunas fixas (que se repetem em cada linha)
    var_name='Year', # Nome da coluna que vai conter os anos
    value_name='Production' # nome da coluna que vai conter a Produção por ano
)
```

```python
# Deletando a coluna Total_production pois ela não está servindo no momento
prod_long.drop('Total_production', axis=1, inplace=True)
```

```python
# prod_long[prod_long['Country'] == 'Brazil']

# Agora podemos agrupar e calcular a média por país
media_producao = prod_long.groupby('Country')['Production'].mean().reset_index()
```

```python
media_producao.head()
```

### Paises com maior exportação total

```python
# Transforma export em formato longo
export_long = df_export.melt(
    id_vars=['Country', 'Total_export'],
    var_name='Year',
    value_name='Export'
)
```

```python
export_long.head()
```

```python
# Soma total exportado por país
total_export = export_long.groupby('Country')['Export'].sum().reset_index()
```

```python
# Ordena do maior para o menor
total_export = total_export.sort_values(by='Export', ascending=False)
```

```python
total_export.head()
```

### Crescimento ao longo dos anos

```python
# Exemplo com o Brasil
brasil = prod_long[prod_long['Country'] == 'Brazil'].sort_values('Year')
```

```python
# Calcula a variação percentual ano a ano
brasil['Growth'] = brasil['Production'].pct_change() * 100
```

```python
brasil['Production'] = brasil['Production'] / 1_000_000  # converte para milhões de sacas/toneladas
```

```python
brasil.head()
```

## Usando Pivot_table

A ideia é responder perguntas como:

“Qual o volume médio de produção por tipo de café ao longo dos anos?”

```python
prod_pivot = pd.pivot_table(
    prod_long,
    values='Production',
    index='Coffee type',  # agrupa por tipo de café
    aggfunc='mean'        # calcula a média de produção
)
```

```python
prod_pivot
```

Outra coisa útil seria ver a média por tipo de café e por ano:

```python
prod_pivot_2 = pd.pivot_table(
    prod_long,
    values='Production',
    index='Year',
    columns='Coffee type',
    aggfunc='mean'
)
```

```python
prod_pivot_2.head()
```

## Usando crosstab

pd.crosstab() serve para contar frequências entre categorias.
No seu caso, podemos perguntar, por exemplo:

“Quantos países produzem cada tipo de café?”

```python
prod_tabela = pd.crosstab(prod_long['Country'], prod_long['Coffee type'])
```
