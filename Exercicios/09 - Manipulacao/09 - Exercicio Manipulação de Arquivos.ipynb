Exercício 1 — CSV básico

Crie um arquivo produtos.csv com:
- nome
- preço

Cadastre 3 produtos.

```python
import csv

filename = 'produtos.csv'

with open(filename, 'w', newline='', encoding='utf8') as arquivo:
  escrever = csv.writer(arquivo)
  escrever.writerow(["nome", "preco"])
  escrever.writerow(["Oleo de Soja", "15.5"])
  escrever.writerow(["Macarrao", "9.50"])
  escrever.writerow(["Feijao", "10.8"])
```

Exercício 2 — Leitura de CSV

Leia o arquivo produtos.csv e:

- mostre apenas os nomes dos produtos
- calcule o preço total

```python
precos_total = 0

with open(filename, 'r', encoding='utf8') as arquivo:
  linhas = csv.reader(arquivo)
  for l in linhas:
    if l[0] != 'nome':
      print(f"Produto: {l[0]}")
      precos_total += float(l[1])

print(f"O preço total dos produtos é R$ {precos_total:.2f}")
```

Exercício 3 — CSV com dicionário

Crie um CSV alunos.csv usando DictWriter com:

- nome
- nota

Depois leia e mostre apenas os alunos com nota ≥ 7.

```python
import csv

with open('alunos.csv', 'w', encoding='utf8', newline='') as arquivo:
  campos = ['nome', 'nota']
  escrever = csv.DictWriter(arquivo, fieldnames=campos)

  escrever.writeheader()
  escrever.writerow({'nome': 'Carlos Silva', 'nota': '7.5'})
  escrever.writerow({'nome': 'Amanda Daniela', 'nota': '9.5'})
  escrever.writerow({'nome': 'Ofelia Castro', 'nota': '10.0'})
  escrever.writerow({'nome': 'Estefanie Lima', 'nota': '4.8'})
```

```python
with open('alunos.csv', 'r', encoding='utf8', newline='') as arquivo:
  linhas = csv.DictReader(arquivo)

  print("Alunos com notas igual ou maior que 7:\n")
  for l in linhas:
    l['nota'] = float(l['nota'])

    if l['nota'] >= 7.0:
      print(f"{l['nome']} : {l['nota']}")
```

Exercício 4 — JSON simples

Crie um arquivo config.json com:

- tema
- idioma
- modo_debug (True ou False)

Depois leia e exiba as configurações.

```python
import json

dados = {
    "tema": "Dark",
    "idioma": "pt-BR",
    "modo_debug": True
}

with open('config.json', 'w', encoding='utf8') as arquivo:
  json.dump(dados, arquivo, ensure_ascii=False, indent=4)
```

```python
with open('config.json', encoding='utf8') as arquivo:
  dados_recebidos = json.load(arquivo)

print(dados_recebidos)
```

Exercício 5 — JSON com lista

Crie um JSON tarefas.json contendo uma lista de tarefas, onde cada tarefa tem:

- descrição
- concluída (True/False)

Depois leia e mostre apenas as tarefas não concluídas.

```python
import json

dados2 = [
    {"descricao": "tarefa1", "concluida": True},
    {"descricao": "tarefa2", "concluida": False},
    {"descricao": "tarefa3", "concluida": False},
    {"descricao": "tarefa4", "concluida": True}
]

with open('tarefas.json', 'w', encoding='utf8') as arquivo:
  json.dump(dados2, arquivo, ensure_ascii=False, indent=4)


with open('tarefas.json', encoding='utf8') as arquivo:
  dados_tarefas = json.load(arquivo)

print("Essas são as tarefas não concluidas")
for d in dados_tarefas:
  if d['concluida'] == False:
    print(f"{d['descricao']} | {d['concluida']}")
```

Exercício 6 — CSV → JSON (intermediário)

Leia um arquivo CSV de alunos e:
- transforme os dados em uma lista de dicionários
- salve tudo em um arquivo JSON

```python
import json
import csv

# ARQUIVO 'alunos.csv' FEITO no EXERCICIO 3

dados3 = []

with open('alunos.csv', 'r', encoding='utf8', newline='') as arquivo:
  linhas = csv.DictReader(arquivo)

  # Transformando os dados em uma lista de dicionarios
  for l in linhas:
    dados3.append(l)


# Salvando dados em formato JSON
with open('alunos2.json', 'w', encoding='utf8') as arquivo:
  json.dump(dados3, arquivo, ensure_ascii=False, indent=4)


# lendo o arquivo json
with open('alunos2.json', encoding='utf8') as arquivo:
  dados_alunos = json.load(arquivo)

for aluno in dados_alunos:
  print(aluno)
```
