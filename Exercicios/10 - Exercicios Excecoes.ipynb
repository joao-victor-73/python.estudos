Trate os erros:

````python
try:
  num1 = float(input("Digite um n√∫mero: "))
  num2 = float(input("Digite outro n√∫mero: "))

  resultado = num1 / num2

except ZeroDivisionError:
  print("A divis√£o n√£o pode ser por 0!")
except ValueError:
  print("Valores digitados n√£o s√£o n√∫meros!")
else:
  print(f'O resultado da divis√£o foi {resultado}')
````

# Exerc√≠cio 3 ‚Äî Acesso a lista
Dada a lista:

> numeros = [10, 20, 30]

Pe√ßa um √≠ndice ao usu√°rio e tente acessar o valor correspondente.


````python
numeros = [10, 20, 30]

try:
  print(f"A lista de numeros √©: {numeros}")
  ind = int(input("Qual indice deseja acessar? "))

  print(f"O valor solicitado foi: {numeros[ind]}")

except IndexError:
  print("Index solicitado √© inexistente!")

except ValueError:
  print("Valor digitado est√° incorreto")
````
